```python
import hashlib

# 1. Робота з текстом
def analyze_text(text):
    words = text.lower().split()
    word_count = {}
    for word in words:
        word = word.strip('.,!?()[]{}"')
        word_count[word] = word_count.get(word, 0) + 1
    frequent_words = [word for word, count in word_count.items() if count > 3]
    return word_count, frequent_words

# 2. Інвентаризація продуктів
inventory = {
    "apple": 10,
    "banana": 3,
    "milk": 2,
    "bread": 7
}

def update_inventory(product, quantity):
    inventory[product] = inventory.get(product, 0) + quantity
    if inventory[product] < 0:
        inventory[product] = 0

def low_stock(threshold=5):
    return [product for product, qty in inventory.items() if qty < threshold]

# 3. Статистика продажів
sales = [
    {"продукт": "apple", "кількість": 20, "ціна": 3},
    {"продукт": "banana", "кількість": 50, "ціна": 2},
    {"продукт": "milk", "кількість": 10, "ціна": 25},
    {"продукт": "bread", "кількість": 40, "ціна": 1.5}
]

def calculate_revenue(sales_data):
    revenue = {}
    for sale in sales_data:
        product = sale["продукт"]
        income = sale["кількість"] * sale["ціна"]
        revenue[product] = revenue.get(product, 0) + income
    high_earning_products = [product for product, amount in revenue.items() if amount > 1000]
    return revenue, high_earning_products

# 4. Система управління задачами
tasks = {
    "Завдання 1": "виконано",
    "Завдання 2": "очікує",
    "Завдання 3": "в процесі"
}

def add_task(name, status="очікує"):
    tasks[name] = status

def remove_task(name):
    if name in tasks:
        del tasks[name]

def change_task_status(name, status):
    if name in tasks:
        tasks[name] = status

def pending_tasks():
    return [name for name, status in tasks.items() if status == "очікує"]

# 5. Аутентифікація користувачів
users = {
    "user1": {
        "password": hashlib.md5("pass123".encode()).hexdigest(),
        "name": "Іван Петренко"
    },
    "admin": {
        "password": hashlib.md5("adminpass".encode()).hexdigest(),
        "name": "Адміністратор Системи"
    }
}

def authenticate_user(login):
    import getpass
    password = input("Введіть пароль: ")
    hashed = hashlib.md5(password.encode()).hexdigest()
    user = users.get(login)
    if user and user["password"] == hashed:
        print(f"Вітаю, {user['name']}! Вхід успішний.")
    else:
        print("Невірний логін або пароль.")


print("=== Завдання 1 ===")
text = "яблуко банан яблуко банан яблуко яблуко хліб молоко хліб молоко"
count, frequent = analyze_text(text)
print("Частота слів:", count)
print("Слова >3 разів:", frequent)

print("\n=== Завдання 2 ===")
update_inventory("banana", 2)
update_inventory("milk", -1)
print("Інвентар:", inventory)
print("Малий залишок:", low_stock())

print("\n=== Завдання 3 ===")
revenue, rich_products = calculate_revenue(sales)
print("Дохід:", revenue)
print("Популярні продукти:", rich_products)

print("\n=== Завдання 4 ===")
add_task("Завдання 4")
change_task_status("Завдання 3", "виконано")
print("Задачі:", tasks)
print("Очікують:", pending_tasks())

print("\n=== Завдання 5 ===")
```
