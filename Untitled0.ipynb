```python
import hashlib

# 1. Робота з текстом
def analyze_text(text):
    words = text.split()
    word_count = {}
    for word in words:
        word = word.lower()
        word_count[word] = word_count.get(word, 0) + 1
    frequent_words = [word for word, count in word_count.items() if count > 3]
    return word_count, frequent_words


# 2. Інвентаризація продуктів
class Inventory:
    def __init__(self):
        self.products = {}

    def update_product(self, name, quantity):
        if quantity == 0:
            self.products.pop(name, None)
        else:
            self.products[name] = quantity

    def low_stock(self):
        return [name for name, qty in self.products.items() if qty < 5]


# 3. Статистика продажів
def sales_statistics(sales_list):
    revenue = {}
    for sale in sales_list:
        product = sale["продукт"]
        quantity = sale["кількість"]
        price = sale["ціна"]
        total = quantity * price
        revenue[product] = revenue.get(product, 0) + total
    profitable = [product for product, income in revenue.items() if income > 1000]
    return revenue, profitable


# 4. Система управління задачами
class TaskManager:
    def __init__(self):
        self.tasks = {}

    def add_task(self, name, status="очікує"):
        self.tasks[name] = status

    def remove_task(self, name):
        self.tasks.pop(name, None)

    def update_status(self, name, status):
        if name in self.tasks:
            self.tasks[name] = status

    def get_pending_tasks(self):
        return [name for name, status in self.tasks.items() if status == "очікує"]


# 5. Аутентифікація користувачів
class UserManager:
    def __init__(self):
        self.users = {}

    def hash_password(self, password):
        return hashlib.md5(password.encode()).hexdigest()

    def add_user(self, login, full_name, password):
        hashed = self.hash_password(password)
        self.users[login] = {"name": full_name, "password": hashed}

    def authenticate(self, login):
        password = input("Введіть пароль: ")
        hashed = self.hash_password(password)
        user = self.users.get(login)
        if user and user["password"] == hashed:
            print("Аутентифікація успішна")
            return True
        else:
            print("Невірний пароль")
            return False
```
