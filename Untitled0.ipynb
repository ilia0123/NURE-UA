```python
import requests
import matplotlib.pyplot as plt
from datetime import datetime

# URL запиту
url = "https://bank.gov.ua/NBU_Exchange/exchange_site?start=20250317&end=20250321&valcode=eur&json"

# Отримання даних з API
response = requests.get(url)
data = response.json()

# Створення словника: дата -> курс
rates = {}
for entry in data:
    date_str = entry["exchangedate"]
    rate = entry["rate"]
    rates[date_str] = rate
    print(f"{date_str}: {rate}")

# Побудова графіка
dates = [datetime.strptime(date, "%d.%m.%Y") for date in rates.keys()]
values = list(rates.values())

plt.figure(figsize=(10, 5))
plt.plot(dates, values, marker='o', linestyle='-', color='blue')
plt.title("Курс євро за тиждень")
plt.xlabel("Дата")
plt.ylabel("Курс (UAH)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
